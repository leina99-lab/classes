# Section 1. if / elif / else

## 개념 설명

```python
if 조건:
    블록1
elif 조건2:
    블록2
else:
    블록3
```

보드게임 평점 분류 기준:
- **명작**: 평점 8.0 이상
- **우수**: 7.0 이상
- **보통**: 6.0 이상
- **낮음**: 6.0 미만

````python
# 개념 시범: 평점 등급 분류
for g in games[:4]:
    r = g['avg_rating']
    if r >= 8.0:    grade = '명작'
    elif r >= 7.0:  grade = '우수'
    elif r >= 6.0:  grade = '보통'
    else:           grade = '낮음'
    print(f'{g["name"]:<20} {r:.2f} → [{grade}]')
````

# Section 2. 비교·논리 연산자

## 개념 설명

| 연산자 | 의미 | 예 |
|:------:|:-----|:---|
| `==` `!=` | 같다/다르다 | `year == 2019` |
| `>` `>=` `<` `<=` | 대소 | `rating >= 8.0` |
| `and` | 둘 다 True | `rating>=7 and votes>50000` |
| `or` | 하나 이상 True | `year<1960 or year>2015` |
| `not` | 반전 | `not (players < 2)` |
| `in` | 포함 | `'Catan' in names` |

````python
# 개념 시범
names = [g['name'] for g in games]
print('Gloomhaven 있음:', 'Gloomhaven' in names)
print('명작이거나 고전(1960년 이전):',
      sum(1 for g in games if g['avg_rating']>=8.0 or g['year']<1960))
````

# Section 4. break / continue / pass

| 키워드 | 동작 |
|:------:|:-----|
| `break` | 즉시 루프 탈출 |
| `continue` | 현재 반복 건너뜀 |
| `pass` | 아무것도 안 함 |

````python
# break vs continue
print('--- break (평점 5 미만 발견 즉시 중단) ---')
for g in games:
    if g['avg_rating'] < 5:
        print(f'중단: {g["name"]}({g["avg_rating"]:.2f})')
        break
    print(f'OK: {g["name"]}')

print('--- continue (평점 6 미만 건너뜀) ---')
for g in games:
    if g['avg_rating'] < 6:
        continue
    print(f'{g["name"]}: {g["avg_rating"]:.2f}')
````

# Section 5. 조건부 표현식 (Ternary)

## 개념 설명
```python
값 = A if 조건 else B
```

````python
# 조건부 표현식으로 등급 리스트
grades = ['명작' if g['avg_rating']>=8 else ('우수' if g['avg_rating']>=7 else '기타')
          for g in games]
for nm, gr in zip([g['name'] for g in games], grades):
    print(f'{nm:<20} [{gr}]')
````

# 인터랙티브: 게임 추천 필터

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

games_data = [
    {'name':'Catan','min_players':3,'max_players':4,'min_time':60,'max_time':120,'year':1995,'avg_rating':7.15,'num_votes':98234},
    {'name':'Pandemic','min_players':2,'max_players':4,'min_time':45,'max_time':75,'year':2008,'avg_rating':7.61,'num_votes':103412},
    {'name':'7 Wonders','min_players':2,'max_players':7,'min_time':30,'max_time':60,'year':2010,'avg_rating':7.77,'num_votes':87312},
    {'name':'Ticket to Ride','min_players':2,'max_players':5,'min_time':45,'max_time':75,'year':2004,'avg_rating':7.43,'num_votes':94501},
    {'name':'Gloomhaven','min_players':1,'max_players':4,'min_time':60,'max_time':120,'year':2017,'avg_rating':8.79,'num_votes':52341},
    {'name':'Azul','min_players':2,'max_players':4,'min_time':30,'max_time':45,'year':2017,'avg_rating':7.87,'num_votes':68901},
    {'name':'Codenames','min_players':2,'max_players':8,'min_time':15,'max_time':30,'year':2015,'avg_rating':7.64,'num_votes':112345},
    {'name':'Wingspan','min_players':1,'max_players':5,'min_time':40,'max_time':70,'year':2019,'avg_rating':8.10,'num_votes':75234},
    {'name':'Clue','min_players':2,'max_players':6,'min_time':45,'max_time':75,'year':1949,'avg_rating':5.84,'num_votes':23456},
    {'name':'Monopoly','min_players':2,'max_players':8,'min_time':60,'max_time':180,'year':1935,'avg_rating':4.37,'num_votes':31234},
]

s_rating  = widgets.FloatSlider(value=7.0,min=1,max=10,step=0.5,description='최소 평점:',
                                 style={'description_width':'initial'})
s_players = widgets.IntSlider(value=2,min=1,max=8,description='플레이어 수:')
s_time    = widgets.IntSlider(value=120,min=15,max=200,step=15,description='최대 시간(분):',
                               style={'description_width':'initial'})
out       = widgets.Output()

def update(change):
    with out:
        clear_output(wait=True)
        filtered=[g for g in games_data
                  if g['avg_rating']>=s_rating.value
                  and g['min_players']<=s_players.value<=g['max_players']
                  and g['max_time']<=s_time.value]
        print(f'조건 충족 게임: {len(filtered)}개')
        for g in sorted(filtered,key=lambda x:x['avg_rating'],reverse=True):
            grade='명작' if g['avg_rating']>=8 else '우수' if g['avg_rating']>=7 else '보통'
            print(f'  [{grade}] {g["name"]:<20} {g["avg_rating"]:.2f} | {g["min_players"]}-{g["max_players"]}명 | {g["max_time"]}분')

for s in [s_rating,s_players,s_time]: s.observe(update,'value')
update(None)
display(widgets.VBox([s_rating,s_players,s_time,out]))
````
