# Section 1. 함수 (def)

## 개념 설명
```python
def 함수명(매개변수, 기본값=값, *args, **kwargs):
    """독스트링"""
    처리
    return 결과
```
함수로 코드를 재사용하고 모듈화한다.

````python
# 개념 시범: 1박 총비용 계산 함수
def calc_total(price, nights, service_rate=0.12):
    """에어비앤비 총비용 = 숙박비 + 서비스 요금"""
    stay_cost    = price * nights
    service_fee  = stay_cost * service_rate
    cleaning_fee = price * 0.1   # 1박 요금의 10%
    total        = stay_cost + service_fee + cleaning_fee
    return {'stay':stay_cost,'service':round(service_fee),'cleaning':round(cleaning_fee),'total':round(total)}

for l in listings[:3]:
    result = calc_total(l['price'], 3)
    print(f'{l["name"][:25]:25s}: ${result["total"]}')
````

# Section 3. 예외처리 (try-except)

## 개념 설명
```python
try:
    위험 코드
except ValueError as e:
    값 오류 처리
except ZeroDivisionError:
    0 나눗셈 처리
except Exception as e:
    모든 예외 처리
else:
    정상 실행 블록
finally:
    항상 실행
```

````python
# 개념 시범: CSV 가격 파싱 안전 처리
raw_prices = ['150', '220N/A', '350', '', '85']
for raw in raw_prices:
    try:
        price = int(raw)
        print(f'파싱 성공: ${price}')
    except ValueError:
        print(f'파싱 실패: "{raw}" → 건너뜀')
````

    try:
        price=int(raw)
        if price<0: raise ValueError('음수 가격 불가')
    except ValueError as e:
        print(f'[오류] {e}: {raw!r}')
        return None
    else:
        print(f'[성공] ${price}')
        return price
    finally:
        print(f'  검증 완료')
for raw in ['150','-50','abc','350']:

````python
import ipywidgets as widgets
from IPython.display import display, clear_output
import math

listings_data = [
    {'id':1,'name':'Cozy Studio Manhattan','neighbourhood':'Manhattan','room_type':'Private room','price':150,'min_nights':2,'reviews':45,'avail':200},
    {'id':2,'name':'Entire Apt Brooklyn','neighbourhood':'Brooklyn','room_type':'Entire home','price':220,'min_nights':3,'reviews':128,'avail':300},
    {'id':3,'name':'Shared Room Harlem','neighbourhood':'Manhattan','room_type':'Shared room','price':45,'min_nights':1,'reviews':12,'avail':100},
    {'id':4,'name':'Private Room Queens','neighbourhood':'Queens','room_type':'Private room','price':85,'min_nights':1,'reviews':77,'avail':250},
    {'id':5,'name':'Luxury Loft Soho','neighbourhood':'Manhattan','room_type':'Entire home','price':350,'min_nights':2,'reviews':200,'avail':180},
    {'id':6,'name':'Budget Room Bronx','neighbourhood':'Bronx','room_type':'Private room','price':55,'min_nights':1,'reviews':30,'avail':320},
    {'id':7,'name':'Williamsburg Modern','neighbourhood':'Brooklyn','room_type':'Entire home','price':175,'min_nights':2,'reviews':95,'avail':280},
    {'id':8,'name':'Central Park View','neighbourhood':'Manhattan','room_type':'Entire home','price':450,'min_nights':3,'reviews':310,'avail':150},
    {'id':9,'name':'Astoria Greek Exp','neighbourhood':'Queens','room_type':'Entire home','price':130,'min_nights':2,'reviews':55,'avail':260},
    {'id':10,'name':'Staten Island','neighbourhood':'Staten Island','room_type':'Entire home','price':200,'min_nights':3,'reviews':20,'avail':100},
]

s_occ  = widgets.FloatSlider(value=0.7,min=0.1,max=1.0,step=0.05,description='점유율:',
                              style={'description_width':'initial'},readout_format='.0%')
s_disc = widgets.IntSlider(value=0,min=0,max=30,description='할인(%):')
rt_sel = widgets.SelectMultiple(options=['Entire home','Private room','Shared room'],
                                 value=['Entire home'],description='방 유형:',rows=3)
out    = widgets.Output()

def update(change):
    with out:
        clear_output(wait=True)
        disc=s_disc.value/100
        occ=s_occ.value
        filtered=[l for l in listings_data if l['room_type'] in rt_sel.value]
        print(f'{"숙소명":<22} {"정가":>6} {"할인가":>7} {"연수익 예상":>12}')
        print('-'*52)
        for l in sorted(filtered,key=lambda x:x['price'],reverse=True):
            discounted=math.ceil(l['price']*(1-disc))
            annual=math.ceil(discounted*l['avail']*occ)
            print(f'{l["name"][:20]:<22} ${l["price"]:>5} ${discounted:>6} ${annual:>10,}')

s_occ.observe(update,'value')
s_disc.observe(update,'value')
rt_sel.observe(update,'value')
update(None)
display(widgets.VBox([s_occ,s_disc,rt_sel,out]))
````
