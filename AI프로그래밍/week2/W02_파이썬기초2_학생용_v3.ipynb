# Section 2. 튜플 (Tuple)

## 개념 설명

**튜플(tuple)**은 순서 있고 **변경 불가(immutable)**한 자료구조이다.

```python
customer = (1, 'Male', 19, 15, 39)  # (ID, 성별, 나이, 수입, 점수)
customer[0]      # 1 (읽기만 가능)
customer[0] = 2  # TypeError! 수정 불가
id, genre, age, income, score = customer  # 언패킹
```
> 변경되면 안 되는 데이터(좌표·등록 정보)에 튜플을 사용한다.

````python
# 고객 정보를 튜플로 저장
cust_tuple = tuple(c.values() for c in [customers[0]])
# (id, genre, age, income, score) 형태로 직접 생성
p = customers[0]
cust_t = (p['id'], p['genre'], p['age'], p['income'], p['score'])
print('튜플:', cust_t)
cid, genre, age, income, score = cust_t
print(f'언패킹: ID={cid}, 성별={genre}, 나이={age}')
````

### Ping 7 — 강사 시범: 튜플 생성·언패킹

### Ping 9 — 강사 시범: 다중 반환값 → 튜플 언패킹

````python
def score_stats(scores):
    s = sorted(scores)
    return min(s), max(s), sum(s)//len(s), s[len(s)//2]

scores = [p['score'] for p in customers]
lo, hi, avg, med = score_stats(scores)
print(f'최소:{lo}, 최대:{hi}, 평균:{avg}, 중앙:{med}')
````

# Section 3. 딕셔너리 (Dictionary)

## 개념 설명

**딕셔너리(dict)**는 키(key)-값(value) 쌍의 자료구조이다.

```python
customer = {'id':1, 'genre':'Male', 'age':19, 'income':15, 'score':39}
customer['age']           # 19 (접근)
customer['vip'] = True   # 키 추가
del customer['vip']       # 키 삭제
customer.get('tier','Basic')  # 없으면 기본값
customer.keys()   # 키 목록
customer.values() # 값 목록
customer.items()  # (키,값) 쌍 목록
```

````python
# 고객 데이터베이스 딕셔너리
db = {p['id']: p for p in customers}
print('고객 5번:', db[5])
print('모든 ID:', list(db.keys())[:5])
````

# Section 4. 집합 (Set)

## 개념 설명

**집합(set)**은 **순서 없고 중복 없는** 자료구조이다.

```python
{1, 2, 3} | {2, 3, 4}   # 합집합 → {1,2,3,4}
{1, 2, 3} & {2, 3, 4}   # 교집합 → {2,3}
{1, 2, 3} - {2, 3, 4}   # 차집합 → {1}
{1,2,3} <= {1,2,3,4}    # 부분집합 → True
```

````python
ages    = {p['age']   for p in customers}
incomes = {p['income'] for p in customers}
print('중복 제거 나이 집합:', sorted(ages))
print('중복 제거 수입 집합:', sorted(incomes))
````

# Section 5. 불리언 (Boolean) & copy()

## 개념 설명

**불리언(bool)** — `True` / `False`
```python
bool(0)   # False   bool(1)   # True
bool('')  # False   bool([])  # False
bool(39)  # True    bool([1]) # True
```

**얕은 복사 vs 깊은 복사**
```python
import copy
a = [1, 2, 3]
b = a           # 같은 객체 참조
c = a.copy()    # 얕은 복사 → 새 객체
d = copy.deepcopy(a)  # 깊은 복사
```

````python
# 불리언 진리값 테스트
print(bool(0), bool(1), bool(''))
print(bool([]), bool([1,2]), bool({}))

# 고객 점수 → 불리언
scores = [p['score'] for p in customers]
is_active = [bool(s) for s in scores]
print('활성 고객:', all(is_active))  # 모두 True?
print('점수 있는 고객 수:', sum(is_active))
````

### 디버그 9. all() vs any()

```python
scores = [39,81,6]
print(all(s>=70 for s in scores))  # 하나라도 70+ 의도
```
오류를 찾아 수정하라.

````python
scores = [39,81,6]
print(all(s>=70 for s in scores))  # 하나라도 70+ 의도
````

<details>
<summary>▶ 디버그 정답 보기</summary>

```python

```

**수정 코드:**
```python
scores = [39,81,6]
print(any(s>=70 for s in scores))  # 하나라도 True이면 True
```

**오류 원인:** all()은 모두 True여야 한다. 하나라도 True이면 되는 경우는 any()를 사용한다.
</details>

# 인터랙티브: 고객 세그먼트 분석기

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

customers_data = [
    {'id':1,'genre':'Male','age':19,'income':15,'score':39},
    {'id':2,'genre':'Male','age':21,'income':15,'score':81},
    {'id':3,'genre':'Female','age':20,'income':16,'score':6},
    {'id':4,'genre':'Female','age':23,'income':16,'score':77},
    {'id':5,'genre':'Female','age':31,'income':17,'score':40},
    {'id':6,'genre':'Female','age':22,'income':17,'score':76},
    {'id':7,'genre':'Female','age':35,'income':18,'score':6},
    {'id':8,'genre':'Female','age':23,'income':18,'score':94},
    {'id':9,'genre':'Male','age':64,'income':19,'score':3},
    {'id':10,'genre':'Female','age':30,'income':19,'score':72},
]

min_score  = widgets.IntSlider(value=0,min=0,max=100,description='최소 소비점수:',
                                style={'description_width':'initial'})
genre_sel  = widgets.SelectMultiple(options=['Male','Female'],value=['Male','Female'],
                                     description='성별:', rows=2)
sort_by    = widgets.Dropdown(options=['score','income','age'],value='score',
                               description='정렬 기준:')
out = widgets.Output()

def update(change):
    with out:
        clear_output(wait=True)
        filtered = [p for p in customers_data
                    if p['score']>=min_score.value and p['genre'] in genre_sel.value]
        sorted_d = sorted(filtered, key=lambda x: x[sort_by.value], reverse=True)
        print(f'필터된 고객: {len(sorted_d)}명')
        print(f'{"ID":^4} {"성별":^7} {"나이":>4} {"수입":>6} {"점수":>5} {"등급":^6}')
        print('-' * 38)
        for p in sorted_d:
            tier = 'VIP' if p['score']>=80 else ('Gold' if p['score']>=50 else 'Basic')
            print(f'{p["id"]:^4} {p["genre"]:^7} {p["age"]:>4}세 {p["income"]:>5}k$ {p["score"]:>5} {tier:^6}')

min_score.observe(update,'value')
genre_sel.observe(update,'value')
sort_by.observe(update,'value')
update(None)
display(widgets.VBox([min_score, genre_sel, sort_by, out]))
````
