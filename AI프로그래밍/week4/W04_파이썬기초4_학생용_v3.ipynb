# Section 3. for + 딕셔너리

## 개념 설명
```python
for k in d:           # 키만
for k,v in d.items(): # 키+값
```

````python
# 개념 시범: 연료 유형별 평균 CO2
fuel_stats={}
for c in cars:
    fuel_stats.setdefault(c['fuel_type'],{'total':0,'count':0})
    fuel_stats[c['fuel_type']]['total']+=c['co2']
    fuel_stats[c['fuel_type']]['count']+=1
for ft,v in fuel_stats.items():
    print(f'연료 {ft}: 평균 CO2 {v["total"]/v["count"]:.1f} g/km')
````

# 인터랙티브: 자동차 연비·CO2 분석기

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

cars_data = [
    {'make':'Toyota','model':'Corolla','year':2022,'engine':1.8,'cyl':4,'fuel_type':'Z','city':8.1,'hwy':6.3,'co2':184},
    {'make':'Honda','model':'Civic','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':8.3,'hwy':6.4,'co2':188},
    {'make':'Mazda','model':'Mazda3','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':8.7,'hwy':7.0,'co2':198},
    {'make':'VW','model':'Golf','year':2022,'engine':1.5,'cyl':4,'fuel_type':'Z','city':8.5,'hwy':7.1,'co2':193},
    {'make':'BMW','model':'3 Series','year':2022,'engine':2.0,'cyl':4,'fuel_type':'X','city':9.1,'hwy':7.0,'co2':208},
    {'make':'Mercedes','model':'C-Class','year':2022,'engine':2.0,'cyl':4,'fuel_type':'X','city':9.5,'hwy':7.3,'co2':215},
    {'make':'Hyundai','model':'Elantra','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':8.5,'hwy':6.7,'co2':193},
    {'make':'Kia','model':'Forte','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':8.7,'hwy':6.9,'co2':197},
    {'make':'Subaru','model':'Impreza','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':9.0,'hwy':7.2,'co2':203},
    {'make':'Nissan','model':'Sentra','year':2022,'engine':2.0,'cyl':4,'fuel_type':'Z','city':7.9,'hwy':6.1,'co2':180},
]

max_co2 = widgets.IntSlider(value=220,min=150,max=250,step=5,description='CO2 상한(g/km):',
                             style={'description_width':'initial'})
fuel_sel= widgets.SelectMultiple(options=['Z','X','D'],value=['Z','X'],description='연료 유형:')
sort_sel= widgets.Dropdown(options=['co2','city','hwy'],value='co2',description='정렬 기준:')
out     = widgets.Output()

def update(change):
    with out:
        clear_output(wait=True)
        filtered=[c for c in cars_data
                  if c['co2']<=max_co2.value and c['fuel_type'] in fuel_sel.value]
        sorted_c=sorted(filtered,key=lambda x:x[sort_sel.value])
        print(f'필터 결과: {len(sorted_c)}대')
        print(f'{"제조사":<10} {"모델":<12} {"도심":>5} {"고속":>5} {"CO2":>5} {"등급":^4}')
        print('-'*46)
        for c in sorted_c:
            g='A' if c['co2']<190 else 'B' if c['co2']<205 else 'C'
            print(f'{c["make"]:<10} {c["model"]:<12} {c["city"]:>5} {c["hwy"]:>5} {c["co2"]:>5} {g:^4}')

max_co2.observe(update,'value')
fuel_sel.observe(update,'value')
sort_sel.observe(update,'value')
update(None)
display(widgets.VBox([max_co2,fuel_sel,sort_sel,out]))
````
