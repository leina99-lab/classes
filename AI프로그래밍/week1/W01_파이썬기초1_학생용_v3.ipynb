# Section 2. 숫자형 연산 (Numeric Operations)

## 개념 설명

| 연산자 | 이름 | 예 | 결과 |
|:------:|:-----|:---|:----:|
| `+` `-` `*` | 사칙 | `145 + 20` | `165` |
| `/` | 실수 나눗셈 | `233 / 2` | `116.5` |
| `//` | 몫 | `233 // 10` | `23` |
| `%` | 나머지 | `63 % 10` | `3` |
| `**` | 제곱 | `2 ** 8` | `256` |

**임상 지표 공식:**
```
심박수 예비능 = 최대심박수 - 안정심박수(약 70)
Framingham 위험점수 ≈ 콜레스테롤/100 + 혈압/10
```

````python
# 개념 *: 환자 1번 임상 지표 계산
age       = 63
trestbps  = 145
chol      = 233
thalach   = 150   # 최대 심박수

# 최대 예상 심박수 공식: 220 - 나이
max_hr_pred  = 220 - age
# 실제 최대심박수 비율
hr_ratio     = thalach / max_hr_pred * 100
# 콜레스테롤 위험 비율
chol_ratio   = chol / 200 * 100

print(f'예상 최대심박수: {max_hr_pred}bpm')
print(f'달성률: {hr_ratio:.1f}%')
print(f'콜레스테롤 지수: {chol_ratio:.1f}% (기준 200 대비)')
````

### Ping 5 —  *: 콜레스테롤 위험 지수 계산

# Section 3. 자료형 변환 (Type Conversion)

## 개념 설명

| 함수 | 역할 | 예 |
|:----:|:-----|:---|
| `int()` | 정수 변환 | `int('63')` → `63` |
| `float()` | 실수 변환 | `float('2.3')` → `2.3` |
| `str()` | 문자열 변환 | `str(145)` → `'145'` |
| `round(x, n)` | 반올림 | `round(2.345, 2)` → `2.35` |

> CSV 파일에서 읽어온 데이터는 모두 **문자열(str)**이다. 연산 전에 반드시 변환해야 한다.

````python
# 개념 *: CSV에서 읽어온 문자열 → 숫자 변환
raw_age  = '63'    # CSV에서 읽어온 값 → str
raw_chol = '233'
raw_peak = '2.3'

age  = int(raw_age)
chol = int(raw_chol)
peak = float(raw_peak)

print(type(raw_age), type(age))    # str → int
print(type(raw_peak), type(peak))  # str → float
print(f'BMI-like: 콜레스테롤/나이 = {chol/age:.2f}')
````

# 인터랙티브: 심장 위험도 계산기

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

s_age  = widgets.IntSlider(value=55,min=30,max=80,description='나이:')
s_bp   = widgets.IntSlider(value=130,min=80,max=200,step=5,description='혈압(mmHg):',
                            style={'description_width':'initial'})
s_chol = widgets.IntSlider(value=220,min=100,max=400,step=5,description='콜레스테롤:',
                            style={'description_width':'initial'})
s_hr   = widgets.IntSlider(value=150,min=60,max=200,description='최대심박수:',
                            style={'description_width':'initial'})
out = widgets.Output()

def update(change):
    with out:
        clear_output(wait=True)
        age, bp, chol, hr = s_age.value, s_bp.value, s_chol.value, s_hr.value
        hr_ratio   = hr / (220 - age) * 100
        chol_level = '고위험' if chol>=240 else ('경계' if chol>=200 else '정상')
        bp_level   = '고혈압' if bp>=140 else ('주의' if bp>=130 else '정상')
        risk = sum([chol>=240, bp>=140, hr_ratio<70, age>=60])
        risk_label = ['낮음','보통','높음','매우 높음'][min(risk,3)]
        print('=' * 35)
        print(f'  나이: {age}세  혈압: {bp}mmHg [{bp_level}]')
        print(f'  콜레스테롤: {chol}mg/dl [{chol_level}]')
        print(f'  심박달성률: {hr_ratio:.1f}%')
        print(f'  종합 위험도: [{risk_label}]')
        print('=' * 35)

for s in [s_age,s_bp,s_chol,s_hr]: s.observe(update,'value')
update(None)
display(widgets.VBox([s_age,s_bp,s_chol,s_hr,out]))
````
