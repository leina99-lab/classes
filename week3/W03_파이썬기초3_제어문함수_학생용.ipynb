## 1. **조건문**(Conditional Statement): if / elif / else

### 1.1 기본 구조

**조건문**은 "만약 ~이면 A를 하고, 아니면 B를 하라"는 분기 로직이다.

````python
if 조건식:
 # 조건이 True일 때 실행
elif 다른조건식:
 # 위 조건이 False이고, 이 조건이 True일 때 실행
else:
 # 모든 조건이 False일 때 실행
````

```
시각화: 조건문 흐름도(Flowchart)

 ┌─────────┐
 │ 시작 │
 └────┬────┘
 ▼
 ┌─────────────┐
 │ cases > 1000 │
 └──┬───────┬──┘
 True │ │ False
 ▼ ▼
 ┌──────────┐ ┌─────────────┐
 │ "위험" │ │cases > 500 │
 └──────────┘ └──┬───────┬──┘
 True │ │ False
 ▼ ▼
 ┌──────────┐ ┌──────────┐
 │ "주의" │ │ "안전" │
 └──────────┘ └──────────┘
```

### 1.2 코로나19 데이터 적용 예시

````python
daily_cases = 1523

if daily_cases > 1000:
 status = "위험"
elif daily_cases > 500:
 status = "주의"
else:
 status = "안전"

print(f"오늘 확진자: {daily_cases:,}명 → 상태: {status}")
````

### 1.3 중요 규칙

- `if` 뒤에 **콜론(`:`)** 을 반드시 붙인다.
- 조건 블록은 **들여쓰기**(Indentation, 보통 스페이스 4칸)로 구분한다.
- 들여쓰기가 맞지 않으면 `IndentationError`가 발생한다.

## 2. **반복문**(Loop): while과 for

### 2.1 **while문** — 조건이 참인 동안 반복

````python
count = 0
while count < 5:
 print(f"반복 {count}")
 count += 1 # 이것을 빼면 무한루프!
````

### 2.2 **for문** — 시퀀스를 순회하며 반복

````python
countries = ["한국", "미국", "일본"]
for country in countries:
 print(country)
````

### 2.3 `range()` 함수

````python
# range(시작, 끝, 간격) — 끝은 포함하지 않음
for i in range(5): # 0, 1, 2, 3, 4
 print(i)

for i in range(1, 10, 2): # 1, 3, 5, 7, 9
 print(i)
````

### 2.4 유용한 반복 도구

| 도구 | 기능 | 예시 |
|---|---|---|
| `enumerate()` | 인덱스와 값을 함께 반환 | `for i, v in enumerate(리스트):` |
| `zip()` | 여러 시퀀스를 병렬로 순회 | `for a, b in zip(리스트1, 리스트2):` |
| `break` | 반복을 즉시 중단 | `if 조건: break` |
| `continue` | 현재 반복을 건너뜀 | `if 조건: continue` |

````python
# enumerate() 예시
daily = [523, 681, 445, 892]
for day, cases in enumerate(daily, start=1):
 print(f"{day}일차: {cases}명")

# zip() 예시
countries = ["한국", "미국", "일본"]
cases = [5000, 85000, 15000]
for country, case in zip(countries, cases):
 print(f"{country}: {case:,}명")
````

### 2.5 `break`와 `continue`

````python
# break: 확진자 1000명 초과하는 첫 날을 찾으면 멈춤
daily = [523, 681, 445, 892, 1023, 756]
for i, cases in enumerate(daily, start=1):
 if cases > 1000:
 print(f"{i}일차에 1000명 초과! ({cases}명)")
 break

# continue: 500명 미만인 날은 건너뜀
for i, cases in enumerate(daily, start=1):
 if cases < 500:
 continue
 print(f"{i}일차: {cases}명 (500명 이상)")
````

## 3. **컴프리헨션**(Comprehension)

### 3.1 **리스트 컴프리헨션**(List Comprehension)

for문을 한 줄로 압축하여 리스트를 생성하는 문법이다.

````python
# 일반 for문
squares = []
for x in range(1, 6):
 squares.append(x ** 2)

# 리스트 컴프리헨션 — 같은 결과를 한 줄로!
squares = [x ** 2 for x in range(1, 6)]
print(squares) # [1, 4, 9, 16, 25]
````

### 3.2 조건부 **컴프리헨션**(Conditional Comprehension)

````python
daily = [523, 681, 445, 892, 1023, 756, 634]

# 평균 이상인 날만 추출
avg = sum(daily) / len(daily)
above_avg = [x for x in daily if x >= avg]
print(above_avg) # [681, 892, 1023, 756]
````

```
시각화: 리스트 컴프리헨션 구조

[ 표현식 for 변수 in 시퀀스 if 조건 ]
 ↑ ↑ ↑
 무엇을 어디서 꺼내 어떤 것만
 만들지 반복할지 걸러낼지
```

### 3.3 **딕셔너리 컴프리헨션**(Dictionary Comprehension)

````python
countries = ["한국", "미국", "일본"]
cases = [5000, 85000, 15000]

# 딕셔너리 컴프리헨션
covid_dict = {k: v for k, v in zip(countries, cases)}
print(covid_dict) # {"한국": 5000, "미국": 85000, "일본": 15000}

# 조건부: 10000명 이상만
big = {k: v for k, v in covid_dict.items() if v >= 10000}
print(big) # {"미국": 85000, "일본": 15000}
````

## 4. **함수**(Function)

### 4.1 왜 함수를 사용하는가?

같은 코드를 여러 번 반복 작성하는 것은 비효율적이고 오류를 유발한다. 함수는 코드를 **재사용 가능한 블록**으로 만드는 것이다.

> 비유: 함수는 "레시피"이다. 한 번 만들어두면, 재료(매개변수)만 바꿔 넣으면 결과(반환값)가 나온다.

### 4.2 함수 정의와 호출

````python
def calculate_avg(data):
 """리스트의 평균을 계산하여 반환하는 함수"""
 return sum(data) / len(data)

# 함수 호출
daily = [523, 681, 445, 892, 1023]
result = calculate_avg(daily)
print(f"평균: {result:.1f}") # 평균: 712.8
````

```
시각화: 함수의 구조

def 함수이름 (매개변수):
 │ │ │
 │ │ └── 입력값 (재료)
 │ └── 함수의 이름
 └── "함수를 정의한다"는 키워드

 return 반환값
 │ │
 │ └── 출력값 (완성된 요리)
 └── "결과를 돌려준다"는 키워드
```

### 4.3 **매개변수**(Parameter)의 종류

````python
# 기본값(Default) 매개변수
def greet(name, greeting="안녕하세요"):
 return f"{greeting}, {name}!"

print(greet("학생")) # 안녕하세요, 학생!
print(greet("학생", "반갑습니다")) # 반갑습니다, 학생!

# 가변 매개변수 (*args)
def total(*numbers):
 return sum(numbers)

print(total(1, 2, 3)) # 6
print(total(1, 2, 3, 4, 5)) # 15

# 키워드 가변 매개변수 (**kwargs)
def print_info(**info):
 for key, value in info.items():
 print(f" {key}: {value}")

print_info(이름="홍길동", 나이=20, 학과="컴퓨터공학")
````

### 4.4 **람다**(Lambda) 함수

한 줄로 정의하는 간단한 익명 함수이다.

````python
# 일반 함수
def double(x):
 return x * 2

# 람다 함수 — 같은 기능
double = lambda x: x * 2

print(double(5)) # 10
````

### 4.5 `map()`과 `filter()`

````python
daily = [523, 681, 445, 892, 1023]

# map: 모든 요소에 함수 적용
thousands = list(map(lambda x: x / 1000, daily))
print(thousands) # [0.523, 0.681, 0.445, 0.892, 1.023]

# filter: 조건에 맞는 요소만 추출
high = list(filter(lambda x: x > 700, daily))
print(high) # [892, 1023]
````

## 5. **예외처리**(Exception Handling)

### 5.1 왜 예외처리가 필요한가?

프로그램 실행 중 예상치 못한 오류가 발생하면 프로그램이 멈춘다. **예외처리**는 오류가 발생해도 프로그램이 계속 실행되도록 하는 안전장치이다.

> 비유: 예외처리는 "안전벨트"이다. 사고(오류)가 나도 치명적인 결과를 막는다.

### 5.2 기본 구조

````python
try:
 # 오류가 발생할 수 있는 코드
 result = 10 / 0
except ZeroDivisionError:
 # 오류 발생 시 실행할 코드
 print("0으로 나눌 수 없습니다!")
except ValueError:
 print("잘못된 값입니다!")
else:
 # 오류가 발생하지 않았을 때 실행
 print(f"결과: {result}")
finally:
 # 오류 여부와 상관없이 항상 실행
 print("계산 종료")
````

```
시각화: try/except 흐름도

 ┌──────────┐
 │ try │
 │ 코드 실행 │
 └──┬────┬──┘
 정상 │ │ 오류 발생
 ▼ ▼
 ┌────────┐ ┌────────────┐
 │ else │ │ except │
 │ 블록 │ │ 오류 처리 │
 └───┬────┘ └─────┬──────┘
 │ │
 └─────┬──────┘
 ▼
 ┌────────────┐
 │ finally │
 │ 항상 실행 │
 └────────────┘
```

### 5.3 흔한 오류 유형

| **오류**(Error) | 원인 | 예시 |
|---|---|---|
| `ZeroDivisionError` | 0으로 나눔 | `10 / 0` |
| `ValueError` | 잘못된 값 변환 | `int("abc")` |
| `IndexError` | 범위 밖 인덱스 | `[1,2,3][5]` |
| `KeyError` | 없는 딕셔너리 키 | `{"a":1}["b"]` |
| `TypeError` | 잘못된 타입 연산 | `"문자" + 3` |
| `FileNotFoundError` | 없는 파일 열기 | `open("없는파일.txt")` |

### 5.4 실전 예시: 사용자 입력 검증

````python
while True:
 try:
 num = int(input("확진자 수를 입력하세요: "))
 if num < 0:
 print("음수는 입력할 수 없습니다.")
 continue
 break
 except ValueError:
 print("숫자만 입력하세요!")

print(f"입력된 확진자 수: {num:,}명")
````

## Ping-Pong ③: **컴프리헨션**(Comprehension) → 조건부 리스트

### Ping (교수자 시연)

### 디버그 ③ 해답

````python
def moving_average(data, window):
 result = []
 for i in range(len(data) - window + 1): # 수정 1: 범위 제한
 segment = data[i:i+window]
 avg = sum(segment) / window
 result.append(avg)
 return result # 수정 2: return 추가

daily = [100, 200, 300, 400, 500, 600, 700]
ma = moving_average(daily, 3)
print(f"이동평균: {ma}")
print(f"원본 길이: {len(daily)}")
print(f"이동평균 길이: {len(ma)}") # 수정 3: ma가 None이 아님
````
